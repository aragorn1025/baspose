```python
import cv2
import numpy as np
import os
import matplotlib.pyplot
import shutil
```

```python
root_dir = 'input/'
input_txt = 'input.txt'
input_png = 'input.png'
```

```python
def show(image):
    matplotlib.pyplot.imshow(image[:,:,::-1])
    matplotlib.pyplot.show()
    print(image.shape)
```

```python
f = open(root_dir + input_txt, 'r')
str_array = f.read().replace('[', '').replace(']', '').replace('\n', '').replace('  ', ' ').split(' ')
for i in range(0, len(str_array)):
    str_array[i] = float(str_array[i])
p = np.reshape(str_array, (25, 3))
print(p)
```

```python
image = cv2.imread(root_dir + input_png)
show(image)
```

```python
def draw_circle(image_in, x, y, name_out = 'output'):
    x, y = int(x), int(y)
    cv2.circle(image, (x, y), 16, (0, 0, 255), 8)
    print('(%d, %d)' % (x, y))
    show(image)
    cv2.imwrite(name_out + '.png', image)

if os.path.isdir('output'):
    shutil.rmtree('output')
os.mkdir('output')
for i in range(0, len(p)):
    image = cv2.imread(root_dir + input_png)
    draw_circle(image, p[i][0], p[i][1], 'output/output-%d' % i)
```
